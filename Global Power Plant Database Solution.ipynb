Importing the MinMaxScaler:
The MinMaxScaler is one of the scaling techniques in sklearn.preprocessing. It rescales the data so that all features are within a specified range (typically 0 to 1).
Creating a Scaler Object:
sc = MinMaxScaler(): This line creates an instance of the MinMaxScaler. By default, it will scale the data to the range [0, 1].
Fitting and Transforming the Data:
x = sc.fit_transform(x): This line first "fits" the scaler to x, determining the minimum and maximum values of each feature. It then "transforms" the data, scaling each feature to the range [0, 1]. This transformation is done by:
Subtracting the minimum value of each feature.
Dividing by the feature's range (maximum - minimum).
Conclusion
Scaling data is a crucial preprocessing step that helps algorithms work more efficiently and effectively. The code snippet shown scales all features in x to the range [0, 1] using MinMaxScaler, which can help ensure a more robust and stable training process.

cross validation

```python
from sklearn.model_selection import cross_val_score

np.random.seed(10)
def rmse_cv(model, x,y):
    rmse =- (cross_val_score(model, x,y, scoring='neg_mean_squared_error', cv=10))
    return(rmse)


models = [LinearRegression(),
             Ridge(),
             SVR(kernel='linear'),
             SVR(kernel='poly'),
             SVR(kernel='rbf'),
             RandomForestRegressor(),
             DecisionTreeRegressor(),  
             GradientBoostingRegressor(),]
            


names = ['LR','R','svr','svr_p','svr_r','RF','DTR','GBR']

for model,name in zip(models,names):
    score = rmse_cv(model,x,y)
    print("{}    : {:.6f}, {:4f}".format(name,score.mean(),score.std()))
```
